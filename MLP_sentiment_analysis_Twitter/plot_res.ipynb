```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
def plot_path(file_path):
    data = pd.read_csv(file_path)
    data_select = data[data["itr"] == -1]

    plt.figure()
    plt.plot(data_select["Epoch"],data_select["testacc"])
    plt.xlim(0, 150)
    plt.xticks(ticks=[0, 50, 100, 150], labels=["0", "50", "100", "150"])
    plt.yticks(ticks=[45, 55, 65, 75], labels=["45", "55", "65", "75"])
    plt.grid(True)
    plt.xlabel("communication round")
    plt.ylabel("accuracy")

    plt.figure()
    plt.plot(data_select["Epoch"],data_select["trainloss"], label = "trainloss")
    plt.plot(data_select["Epoch"],data_select["testloss"], label = "testloss")
    plt.xlim(0, 150)
    plt.xlabel("communication round")
    plt.ylabel("trainloss")
    plt.grid(True)
    plt.legend()

def plot_folder(folder_path):
    fig, axs = plt.subplots(1, 2, figsize=(8, 4))
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        data = pd.read_csv(file_path)
        data_select = data[data["itr"] == -1]
        axs[0].plot(data_select["Epoch"],data_select["testacc"], label = file_name.split("_")[0])
        axs[0].legend()
        axs[0].set_xlabel("communication round")
        axs[0].set_ylabel("test accuracy (%)")
        axs[0].set_xticks(ticks=[0, 50, 100, 150], labels=["0", "50", "100", "150"])
        axs[0].set_yticks(ticks=[45, 55, 65, 75], labels=["45", "55", "65", "75"])
        axs[0].grid(True)

        axs[1].plot(data_select["Epoch"],data_select["trainloss"], label = file_name.split("_")[0])
        axs[1].legend()
        axs[1].set_xlabel("communication round")
        axs[1].set_ylabel("train loss")
```

## Plot the loss and accuracy curve for local training.

```python
file_path = "/home/peng/workspace/Power-of-Choice/MLP_sentiment_analysis_Twitter/logs/default/lr0.0500_bs32_cp100_a0.20_e1_r0_n314_f0.10/rand_rr0.10_dr0.75_lr0.050_bs32_cp100_a0.20_e1_r0_n314_f0.10_p32.csv"
data = pd.read_csv(file_path)
data_select = data[data["itr"] != -1]

x_axis = np.array([i for i in range(len(data_select))])

N = 20


plt.figure(figsize=(10, 2))
plt.plot(x_axis[0:-N+1]/100, np.convolve(data_select["loss"], np.ones(N)/N, mode='valid'))
plt.ylabel("train loss")
plt.xlabel("client count")
plt.xlim(100,120)

plt.figure(figsize=(10, 2))
plt.plot(x_axis[0:-N+1]/100, np.convolve(data_select["Prec@1"], np.ones(N)/N, mode='valid'))
plt.xlim(100,120)
plt.xlabel("client count")
plt.ylabel("Accu")
```
